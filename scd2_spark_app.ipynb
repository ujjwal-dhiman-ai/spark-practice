```python
CHECKPOINT_EVERY = 2
```

```python
processed = 4
```

```python
if processed % CHECKPOINT_EVERY == 0:
    print(f"Processed {processed} records, checkpointing...")
```

```python
for processed in range(0, 721):
    if processed % CHECKPOINT_EVERY == 0:
        print(f"Processed {processed} records, checkpointing...")
        processed+=1
```

```python
num_snapshots = 300
if num_snapshots <= 5:
    CHECKPOINT_EVERY = num_snapshots  # checkpoint only at the end
else:
    CHECKPOINT_EVERY = max(10, num_snapshots // 4)  # every ~20%

CHECKPOINT_EVERY
```
